## **1. Checking for null values: No null values found**


```python
missing_superstore_df = superstore_df.isnull().sum().to_frame().rename(columns={0:"Total No. of Missing Values"})

# Divided the total No. of Missing values by the Total length of the dataframe, then round it to 2 decimal places
missing_superstore_df["% of Missing Values"] = round((missing_superstore_df["Total No. of Missing Values"]/len(superstore_df))*100,2)



missing_superstore_df
```

## Visualize the Duplicates:

```python
duplicates_count = superstore_df.duplicated().sum()
non_duplicates_count = len(superstore_df) - duplicates_count



# Create data for the pie chart
labels = ['Duplicates', 'Non-Duplicates']
sizes = [duplicates_count, non_duplicates_count]
colors = ['red', 'green']
explode = (0.1, 0)


# Create the pie chart
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=140)
plt.title('Visualize the Duplicates')



plt.show()
print(f"Number of Duplicates found: {duplicates_count}")
print(f"Number of Non-Duplicates: {non_duplicates_count}")
print(f"Superstore Dataset: {len(superstore_df)}")
```

## **Numerical Descriptive Statistics** 
### **(with the Five-number Summary)**

- Count: It shows the number of values that are not missing in each of the column.<br>
- Mean : It shows the average of the values in each of the column.<br>
- Std :  It shows standard deviation of the values in each of the column.<br>
- Min :  It shows the smallest value in each of the column.<br>
- Max :  It shows the maximum value in each of the column.<br>
- 25%, 50%, 75%: The valuescorrespond to the data divisions of 25%, 50% and 75% percentiles.<br></div>

# **1. Ship Mode**

```python
# value_counts() = Similar to COUNT() on SQL

superstore_df['Ship Mode'].value_counts()
```

```python
# Create a figure with three subplots side by side
plt.figure(figsize=(15, 5))


# First subplot: Bar chart for top ship modes
top_ship_modes = superstore_df['Ship Mode'].value_counts().nlargest(30)

plt.subplot(1, 3, 1)
top_ship_modes.plot(kind='bar', color='Green', width=0.5)
plt.title('Top Ship Modes by Count')
plt.xticks(rotation=45)



# Second subplot: Boxplot of Profit by Ship Mode
plt.subplot(1, 3, 2)
sns.boxplot(x='Ship Mode', y='Profit', data=superstore_df, palette='Blues')
plt.title('Boxplot of Profit by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.xticks(rotation=45)



# Third subplot: Pie chart for distribution of ship modes
ship_mode_counts = superstore_df['Ship Mode'].value_counts()

plt.subplot(1, 3, 3)
plt.pie(ship_mode_counts,
        labels=ship_mode_counts.index,
        autopct='%1.1f%%',
        startangle=140,
        colors=['lightblue', 'tomato', 'lightgreen', 'Yellow'])
plt.title('Distribution of Ship Modes (COUNT)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle




plt.tight_layout()
plt.show()
print(ship_mode_counts)
```

### Inferences:

- We make the MOST and LEAST profit on Standard class Ship Mode.

- It contains the 2 biggest outliers (-6,000 and 8,000)

- Standard class also has 59.7% of the total share

## Stacked Bar Chart - Ship Mode and Segment

```python
cross_tab = pd.crosstab(superstore_df['Segment'], superstore_df['Ship Mode'])


plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True, colormap='viridis')


plt.title('Stacked Bar Chart of Ship Mode by Segment')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.xticks(rotation=0)  
plt.legend(title='Ship Mode', loc='upper right')


plt.show()
print(cross_tab)
```

In this code:

1. We use `pd.crosstab` from the pandas library to create a contingency table that shows the count of each combination of "Segment" and "Ship Mode."

2. We create a stacked bar chart by setting `stacked=True` when calling the `plot()` method on the `cross_tab` DataFrame.






## Heatmap - Region vs Category

In this code:

1. We first use `pd.crosstab` to create a cross-tabulation or contingency table of "Region" and "Category." This table shows the count of each combination of region and category.

2. We create a heatmap using Seaborn's `heatmap()` function. We provide the cross-tabulation as the data, set `annot=True` to display the cell values, use `fmt='d'` to format cell values as integers, and choose a colormap (`cmap`) for the heatmap.
